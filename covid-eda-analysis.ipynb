```python
import pandas as pd

import numpy as np

import matplotlib.pyplot as plt

import seaborn as sns
```

```python
#Load Covid DataSet

df=pd.read_csv('Covid Data.csv')

#Display fist 5 rows

print("First 5 rows of dataset:")

print(df.head())
```

```python
#Dataset information

print("\nDataset Info:")

print(df.info())
```

```python
#Missing Values

print("\nMissing Values:")

print(df.isnull().sum())
```

```python
#Statistical summary

print("\nStatistical Summary:")

print(df.describe())
```

```python
#Univariate Analysis

sns.histplot(df ['AGE'], kde=True)

plt.title("Age Distribution")

plt.show()

sns.countplot(x='DIABETES', data=df)

plt.title("Diabetes Count")

plt.show()
```

```python
#Bivariate Analysis

sns.boxplot(x='CLASIFFICATION_FINAL', y='AGE', data=df)

plt.show()
```

```python
#Multi variate analysis 
plt.figure(figsize=(18,16))

corr_matrix=df.select_dtypes(include=['int64','float64']).corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap")

plt.show()
```

```python
#Outliers Detection

sns.boxplot(x=df['AGE'])

plt.title("Age Outliers")

plt.show()
```

```python
#Key Insights

print("Key Insights:")

print("Most patients are in the age group 50-70")

print("Higher mortality seen in elderly patients")

print("Patients with diabetes & hypertension show higher risk")

print("Strong relationship between pneumonia and death")
```
